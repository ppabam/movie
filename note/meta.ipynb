```python
PATH = "~/data/movies/dailyboxoffice/merge/dt=20240102"
new_df = pd.read_parquet(PATH)[['movieCd','multiMovieYn', 'repNationCd']]
len(new_df)
```

```python
old_df =  pd.read_parquet('~/temp/meta/meta.parquet')
```

```python
# movieCd 기준으로 outer join
merged_df = old_df.merge(new_df, on="movieCd", how="outer", suffixes=("_old", "_new"))
merged_df
```

```python
# NaN 값 대체 (A 값 우선, 없으면 B 값 사용)
merged_df["multiMovieYn"] = merged_df["multiMovieYn_old"].combine_first(merged_df["multiMovieYn_new"])
merged_df["repNationCd"] = merged_df["repNationCd_old"].combine_first(merged_df["repNationCd_new"])

# 불필요한 컬럼 제거
merged_df = merged_df[["movieCd", "multiMovieYn", "repNationCd"]]

# 결과 출력
print(merged_df)
```

```python
import pandas as pd

# 예제 데이터
A = pd.DataFrame({
    "movieCd": ["1001", "1002", "1003"],
    "multiMovieYn": ["Y", "Y", "N"],
    "repNationCd": ["K", "F", None]
})

print(A)

B = pd.DataFrame({
    "movieCd": ["1001", "1003", "1004"],
    "multiMovieYn": [None, "Y", "Y"],
    "repNationCd": [None, "F", "K"]
})

print(B)

# movieCd 기준으로 outer join
merged_df = A.merge(B, on="movieCd", how="outer", suffixes=("_A", "_B"))
print(merged_df)

# NaN 값 대체 (A 값 우선, 없으면 B 값 사용)
merged_df["multiMovieYn"] = merged_df["multiMovieYn_A"].combine_first(merged_df["multiMovieYn_B"])
merged_df["repNationCd"] = merged_df["repNationCd_A"].combine_first(merged_df["repNationCd_B"])

# 불필요한 컬럼 제거
merged_df = merged_df[["movieCd", "multiMovieYn", "repNationCd"]]

# 결과 출력
print(merged_df)
```
